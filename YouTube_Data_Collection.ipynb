```python
def get_youtube_data(api_key, max_results=5000):
    youtube = googleapiclient.discovery.build("youtube", "v3", developerKey=api_key)

    categories = ['1','2','10','17','24','25','28']
    all_data = []

    for category_id in categories:
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            chart="mostPopular",
            regionCode="US",
            videoCategoryId=category_id,
            maxResults=50
        )

        response = request.execute()
        results = response.get('items', [])

        # X·ª≠ l√Ω ph√¢n trang
        while 'nextPageToken' in response and len(results) < max_results:
            time.sleep(1)  # Ch·ªù 1 gi√¢y gi·ªØa c√°c y√™u c·∫ßu ƒë·ªÉ tr√°nh b·ªã ch·∫∑n
            request = youtube.videos().list(
                part="snippet,contentDetails,statistics",
                chart="mostPopular",
                regionCode="US",
                videoCategoryId=category_id,
                maxResults=50,
                pageToken=response['nextPageToken']
            )
            response = request.execute()
            results.extend(response.get('items', []))

            if len(results) >= max_results:
                break

        for item in results:
            video_data = {
                'video_id'      : item['id'],
                'title'         : item['snippet']['title'],
                'description'   : item['snippet']['description'],
                'published_at'  : item['snippet']['publishedAt'],
                'channel_id'    : item['snippet']['channelId'],
                'channel_title' : item['snippet']['channelTitle'],
                'category_id'   : category_id,
                'tags'          : item['snippet'].get('tags', []),
                'duration'      : item['contentDetails']['duration'],
                'definition'    : item['contentDetails']['definition'],
                'caption'       : item['contentDetails'].get('caption', 'false'),
                'view_count'    : item['statistics'].get('viewCount', 0),
                'like_count'    : item['statistics'].get('likeCount', 0),
                'dislike_count' : item['statistics'].get('dislikeCount', 0),
                'favorite_count': item['statistics'].get('favoriteCount', 0),
                'comment_count' : item['statistics'].get('commentCount', 0),
            }
            all_data.append(video_data)

    df = pd.DataFrame(all_data)
    return df

# L∆∞u v√†o file CSV
def save_to_csv(data, filename):
  data.to_csv(filename, index=False)
  print(f"D·ªØ li·ªáu ƒë√£ ƒë∆∞·ª£c l∆∞u v√†o {filename}")

def main():
  youtube_data = get_youtube_data(api_key, max_results=5000)
  filename = 'youtube_data.csv'
  save_to_csv(youtube_data, filename)

```

```python
if __name__ == "__main__":
  main()
```
